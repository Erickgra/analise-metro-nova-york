````python
import pandas
import numpy as np

def get_hourly_entries(df):
    
    
    #cria uma coluna nova (ENTRIESn_hourly) no dataframe e preenche suas linhas com a diferença de entradas
    # da linha atual com a linha anterior da coluna (ENTRIESn)
    df['ENTRIESn_hourly'] = df['ENTRIESn'] - df['ENTRIESn'].shift(1)
    #Substitui as linhas com valor == NaN por 1
    df.fillna(1, inplace=True)
    #converte o valor da coluna de float para int
    df['ENTRIESn_hourly'] = df['ENTRIESn_hourly'].apply(np.int64)
    return df

turnstile_regular = get_hourly_entries(turnstile_regular)
turnstile_regular.head()
````

### *Exercicio 2.3*

Calcule se estava nebuloso ou não (0 ou 1) e a temperatura máxima para fog (isto é, a temperatura máxima 
     para dias nebulosos).

````python

def max_temp_aggregate_by_fog(df):
    #agrupa os dados pela data
    df = df.groupby('DATEn').max()
    #recupera o total de dias que estava nebuloso
    fog_day = df['fog'][df['fog'] >= 1].count()
    #pega a maior temperatura de um dia nebuloso
    maxtempi_fog_day = df['maxtempi'][df['fog'] >= 1].max()
    #recupera o total de dias que não estava nebuloso
    non_fog_day = df['fog'][df['fog'] == 0].count()
    #pega a maior temperatura de um dia que não está nebuloso
    maxtempi_non_fog_day = df['maxtempi'][df['fog'] == 0].max()
    return fog_day, maxtempi_fog_day, non_fog_day, maxtempi_non_fog_day
    

print(max_temp_aggregate_by_fog(turnstile_with_weather))
````

### *Exercicio 2.4

Calcule agora a média de 'meantempi' nos dias que são sábado ou domingo (finais de semana):

````python
import csv

from datetime import datetime
#retorna um Date a partir de uma string no formado yyy-MM-dd
def to_date(dateStr):
    return datetime.strptime(dateStr, '%Y-%m-%d')

#retorna o dia da semana de uma dada
def get_weekday(date):   
    return int(date.strftime('%w'))

#verifica se a data é fim de semana
def is_weekend(weekday):
    return (weekday == 0 or weekday == 6)

#retorna a média da temperatura nos fins de semana
def avg_weekend_temperature(filename):
    turnstile_with_weather = pd.read_csv(filename)
    
    with open(filename, 'r') as f_in:
        in_reader = csv.DictReader(f_in)
        total = 0
        count = 0
        previous_date = None
        for row in in_reader:
            data = to_date(row['DATEn'])
            if is_weekend(get_weekday(data)) and (data != previous_date):
                total += float(row['meantempi'])
                count += 1
            previous_date = data
        print(count)
    return total / count

filename = "turnstile_data_master_with_weather.csv"
avg_weekend_temperature(filename)
````

### *Exercicio 2.5

Calcule a média da temperatura mínima 'mintempi' nos dias chuvosos onde da temperatura mínima foi maior que do 55 graus:

````python
# Verifica se é um dia chuvoso
def is_rainy_day(rain):
    return int(float(rain)) >= 1

def avg_min_temperature(filename):
    turnstile_with_weather = pd.read_csv(filename)
    with open(filename, 'r') as f_in:
        in_reader = csv.DictReader(f_in)
        total = 0
        count = 0
        previous_date = None
        for row in in_reader:
            data = to_date(row['DATEn'])
            temp = float(row['mintempi'])
            if is_rainy_day(row['rain']) and temp > 55 and (data != previous_date):
                total += temp
                count += 1
            previous_date = data
        print(count)
    return total / count

filename = "turnstile_data_master_with_weather.csv"
avg_min_temperature(filename)
````

Agora crie o redutor. Dado o resultado do mapeador do exercicio anterior, o redutor deve imprimir(Não retornar) uma linha por UNIT, juntamente com o número total de ENTRIESn_hourly.Ao longo de maio (que é a duração dos nossos dados), separados por uma guia. Um exemplo de linha de saída do redutor pode ser assim: 'R001 \ t500625.0'

Você pode assumir que a entrada para o redutor está ordenada de tal forma que todas as linhas correspondentes a uma unidade particular são agrupados. No entanto a saida do redutor terá repetição pois existem lojas que aparecem em locais diferentes dos arquivos.

Exporte seu redutor em um arquivo chamado reducer_result.txt e envie esse arquivo juntamente com a sua submissão.

````python
def reducer():
    oldKey = None
    total = 0
    for line in sys.stdin:
        data_mapped = line.strip().split(' ')
        if len(data_mapped) != 2:
            continue
        unit, entries = data_mapped
        if 'UNIT' == unit:
            continue
        
        if oldKey and oldKey != unit:
            print(oldKey, "\t", total)
            total = 0
            oldKey = unit
        
        total += float(entries)
        oldKey = unit
        
    if oldKey != None:
        print(oldKey, "\t", total)


if __name__ == "__main__":     
    sys.stdin = open('mapper_result.txt')
    sys.stdout = open('reducer_result.txt', 'w')
    reducer()
````
